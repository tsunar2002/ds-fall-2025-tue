### Clean the Data

Perhaps we‚Äôd like to know more another day, but the team is really interested in just the relationship between calories (`calories_day`) and weight. ‚Ä¶and maybe gender.

Can you remove the other columns?

```python
# Remove ‚Äòem.
cols_to_keep = ['calories_day', 'weight', 'Gender']
# df_food.drop(~cols_to_keep, axis=1, inplace=True)   # doesn't work

cols_to_remove = df_food.columns.difference(cols_to_keep)
df_food.drop(cols_to_remove, axis=1, inplace=True)
  # axis=1 : apply column-wise
  # inplace=True : work on this df

df_food.head()
```

What about `NaN`s? How many are there?

```python
# Count ‚Äòem.
df_food.isna().sum()
```

We gotta remove those `NaN`s‚Äïthe entire row.

```python
# Drop ‚Äòem.
df_food.dropna(inplace=True)
df_food.head()
```

But what about the weird non-numeric values in the column obviously meant for numeric data?

Notice the data type of that column from when you got the types of all the columns?

If only we could convert the column to a numeric type and drop the rows with invalid values. ü§î

```python
# Fix that.

## convert to numeric dtype, "If ‚Äòcoerce‚Äô, then invalid parsing will be set as NaN."
df_food['weight'] = pd.to_numeric(df_food['weight'], errors='coerce')

# drop NaN values
df_food.dropna(subset=['weight'], inplace=True)

df_food.head()
```

Now this data seems reasonably clean for our purposes! üòÅ

### Clean the Data

Well, we can‚Äôt get our answers with what we currently have, so you‚Äôll have to make some changes.

Let‚Äôs not worry about anything below the first 5 values for now. Convert the top 5 to a single canonical value‚Äïsay, `'US'`, which is nice and short.

```python
# Replace them all with 'US'.

# will standardize to: United States (then US) -- used (inefficient) BRUTEFORCE approach, but regex probably handles most (if not all) of these cases
usa_rename = {
    'United States' : 'United States',
    'USA' : 'United States',
    'United States of America' : 'United States',
    'US' : 'United States',
    'U.S.' : 'United States',
    'Usa' : 'United States',
    'usa' : 'United States',
    'United states' : 'United States',
    'united states' : 'United States',
    'Us' : 'United States',
    'us' : 'United States',
    'U.S.A' : 'United States',
    'U.S.A.' : 'United States',
    'United State of America' : 'United States',
    'UnitedStates' : 'United States',
    'U.S' : 'United States',
    'United States Of America' : 'United States',
    'united States' : 'United States',
    'United States of america' : 'United States',
    'U.s.' : 'United States',
    'America' : 'United States',
    'Unite States' : 'United States',
    'United Stateds' : 'United States',
    'Uniyed states' : 'United States',
    'United State' : 'United States',
    'United Stated' : 'United States',
    'ISA' : 'United States',            # perhaps it would've been better to replace invalid values to nulls? -- typos/errors like this could get messy/confusing
    'Isa' : 'United States',
    'Puerto Rico' : 'United States',    # technically, it's a US territory (not its own country, potential 51st state)
    'United STates' : 'United States',
    'Unites States' : 'United States',
    'United Stares' : 'United States',
    'San Francisco' : 'United States',  # state in US
}

us_rename = { 'United States' : 'US' }

# select all rows in country col: strip whitespace, then replace values using dict
##  df_usd_tech['country'] = df_usd_tech['country'].str.strip().str.title().replace(usa_rename)  ->  Try using .loc[row_indexer,col_indexer] = value instead
df_usd_tech.loc[:, 'country'] = df_usd_tech['country'].str.strip().replace(usa_rename).replace(us_rename)

# sanity check
# df_usd_tech['country'].value_counts()
```

Have a look at the count of each unique country again now.

```python
# there are slight issues for other countries as well...

df_usd_tech['country'].value_counts()
```

```python
# BONUS CREDIT: resolve [most of] those anomalous cases too without exhaustively taking every variant literally into account.

# title case (helps standardize other country typos, like Singapore), regex Denmark (alt to mapping/dict)
df_usd_tech.loc[1:, 'country'] = df_usd_tech['country'].str.title().replace(usa_rename).replace(us_rename)
df_usd_tech.loc[ :, 'country'] = df_usd_tech['country'].str.replace(r'^(Denmark|Danmark|denmark)$', 'Denmark', case=False, regex=True)
df_usd_tech['country'] = df_usd_tech['country']

# df_usd_tech['country'].value_counts()
```
